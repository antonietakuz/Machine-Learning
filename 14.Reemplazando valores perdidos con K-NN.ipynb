## Reemplazando valores perdidos con K-NN

```python
Nota: si quisieramos utilizar este algoritmo con variables categoricas, primero debemos convertirlas a numéricas
```

#### 1. Carga de librerias y datos

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from matplotlib import rcParams

from sklearn.impute import KNNImputer
```

```python
niveles=pd.read_csv('C:/Users/pc/Desktop/rociochavez/Rocio-Chavez-youtube-Files-master/Niveles de contaminacion.csv')
```

```python
niveles
```

#### 2. Reemplazo utilizando solo un vecino

```python
#creamos el modelo para llevar adelante la imputación

imputacion = KNNImputer (n_neighbors=1)
```

```python
imputacion.n_neighbors
```

```python
imputacion.metric
```

#### 3. Aplicamos el modelo de imputación al objeto niveles

```python
niveles_knn=imputacion.fit_transform(niveles)
```

```python
niveles_knn
```

##### 4. Convertimos el array en un dataframe

```python
niveles_knn=pd.DataFrame(niveles_knn)
niveles_knn
```

#### Le asignamos los nombres a las columnas del data frame que creamos

```python
niveles_knn.columns= list(niveles.columns.values)
niveles_knn
```
