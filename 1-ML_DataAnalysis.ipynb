### Missing values

Let's go ahead and find out which variables of the dataset contain missing values.

```python
# se hace una lista de los valores missings
vars_with_na = [var for var in data.columns if data[var].isnull().sum() > 0]

# determinamos el porcentaje de missing values
data[vars_with_na].isnull().mean()
```

Our dataset contains a few variables with missing values. We need to account for this in our following notebook / video, where we will engineer the variables for use in Machine Learning Models.

#### Relación entre los valores que faltan y el precio de venta

Evaluemos el precio de la casa en aquellas observaciones donde falta la información, para cada variable.

```python
def analyse_na_value(df, var):

    df = df.copy()

    # hagamos una variable que indique 1 si falta la observación o cero en caso contrario
    df[var] = np.where(df[var].isnull(), 1, 0)

    # comparemos la mediana de SalePrice en las observaciones donde faltan datos
    # vs las observaciones donde hay un valor disponible

    df.groupby(var)['SalePrice'].median().plot.bar()

    plt.title(var)
    plt.show()


# ejecutemos la función en cada variable con datos faltantes
for var in vars_with_na:
    analyse_na_value(data, var)
```

El Precio de Venta promedio en casas donde falta la información, difiere del Precio de Venta promedio en casas donde existe información.

### Variables numéricas

Avancemos y averigüemos qué variables numéricas tenemos en el conjunto de datos.

```python
# lista de variables numericas
num_vars = [var for var in data.columns if data[var].dtypes != 'object']

print('Number of numerical variables: ', len(num_vars))

# visualizar las variables numericas
data[num_vars].head()
```

La variable Id, que es un indicador de la casa. No usaremos esta variable para hacer nuestras predicciones

#### Temporal variables

We have 4 year variables in the dataset:

- YearBuilt
- YearRemodAdd
- GarageYrBlt
- YrSold

Por lo general, no usamos variables de fecha en su formato original. En su lugar, extraemos información de ellos. Por ejemplo, podemos capturar la diferencia de años entre el año en que se construyó la casa y el año en que se vendió.

```python
# listamos las variables que contienen informacion de años

year_vars = [var for var in num_vars if 'Yr' in var or 'Year' in var]

year_vars
```

```python
# Exploramos los valores de las variables que son de tiempo

for var in year_vars:
    print(var, data[var].unique())
    print()
```

Como era de esperar, los valores son años.

Podemos explorar la evolución del precio de venta con los años en que se vendió la casa:

```python
data.groupby('YrSold')['SalePrice'].median().plot()
plt.ylabel('Median House Price')
plt.title('Change in House price with the years')
```

Ha habido una caída en el valor de las casas. Eso es inusual, en la vida real, los precios de la vivienda suelen subir con el paso de los años.

Avancemos y exploremos si existe una relación entre las variables del año y SalePrice. Para ello capturaremos los años transcurridos entre las variables Año y el año en que se vendió la casa:

```python
# exploremos la relación entre las variables del año
# y el precio de la casa con un poco más de detalle:

def analyse_year_vars(df, var):
    df = df.copy()
    
    # capturar la diferencia entre la variable del año y el año
    # en que se vendió la casa
    df[var] = df['YrSold'] - df[var]
    
    plt.scatter(df[var], df['SalePrice'])
    plt.ylabel('SalePrice')
    plt.xlabel(var)
    plt.show()
    
    
for var in year_vars:
    if var !='YrSold':
        analyse_year_vars(data, var)
    
```

Hoy una tendencia de decremento en precio con caracteristicas viejas. Mayor es el tiempo entre que la casa fue construida o remodelada y el año de venta

#### Variables discretas
Avancemos y encontremos qué variables son discretas, es decir, mostrar un número finito de valores

```python
#  Lisamos las varibles discretas 
discrete_vars = [var for var in num_vars if len(
    data[var].unique()) < 20 and var not in year_vars+['Id']]


print('Number of discrete variables: ', len(discrete_vars))
```

```python
# Visualizamos las variables discretas

data[discrete_vars].head()
```

Estas variables discretas tienden a ser calificaciones o escalas de calificación, o se refieren al número de habitaciones o unidades.

Avancemos y analicemos su contribución al precio de la vivienda.

```python
def analyse_discrete(df, var):
    df = df.copy()
    df.groupby(var)['SalePrice'].median().plot.bar()
    plt.title(var)
    plt.ylabel('Median SalePrice')
    plt.show()
    
for var in discrete_vars:
    analyse_discrete(data, var)
```

Suele haber una relación entre los valores de las variables y el precio de venta, pero esta relación no siempre es monótona.

Por ejemplo, para OverallQual, existe una relación monotónica: a mayor calidad, mayor precio de venta.

Sin embargo, para OverallCond, la relación no es monótona. Claramente, algunos grados de Condición, como 5, se correlacionan con precios de venta más altos, pero los valores más altos no necesariamente lo hacen. Debemos tener cuidado con la forma en que diseñamos estas variables para extraer el máximo valor para un modelo lineal.


#### Variables continuas

Avancemos y encontremos la distribución de las variables continuas. 
Consideraremos variables continuas a todas aquellas que no sean variables temporales o discretas en nuestro conjunto de datos.

```python
# make list of continuous variables
cont_vars = [
    var for var in num_vars if var not in discrete_vars+year_vars+['Id']]

print('Number of continuous variables: ', len(cont_vars))
```

```python
# let's visualise the continuous variables

data[cont_vars].head()
```

```python
# Analizamos las distribuciones de estas variables


def analyse_continuous(df, var):
    df = df.copy()
    df[var].hist(bins=30)
    plt.ylabel('Number of houses')
    plt.xlabel(var)
    plt.title(var)
    plt.show()


for var in cont_vars:
    analyse_continuous(data, var)
```

Las variables no se distribuyen normalmente, incluida la variable de destino 'SalePrice'.

Para maximizar el rendimiento de los modelos lineales, debemos tener en cuenta las distribuciones no gaussianas.

Evaluemos si una transformación logarítmica de las variables arroja valores que siguen una distribución normal:

```python
# Avancemos y analicemos las distribuciones de estas variables
# después de aplicar una transformación logarítmica


def analyse_transformed_continuous(df, var):
    df = df.copy()

   # log no toma 0 o valores negativos, así que seamos
    # cuidado y omite esas variables
    if any(data[var] <= 0):
        pass
    else:
        # log transform the variable
        df[var] = np.log(df[var])
        df[var].hist(bins=30)
        plt.ylabel('Number of houses')
        plt.xlabel(var)
        plt.title(var)
        plt.show()


for var in cont_vars:
    analyse_transformed_continuous(data, var)
```

Obtenemos una mejor distribución de los valores para la mayoría de las variables cuando usamos la transformación logarítmica. Es muy probable que este paso de ingeniería agregue valor de rendimiento a nuestro modelo final.

```python
# exploremos la relación entre el precio de la vivienda y
# las variables transformadas con más detalle:


def transform_analyse_continuous(df, var):
    df = df.copy()

    # log no toma valores negativos, así que tengamos cuidado y saltemos esas variables
    if any(data[var] <= 0):
        pass
    else:
        # log transforma la variable
        df[var] = np.log(df[var])
        
        # log transforma el objetivo (recuerda que también estaba sesgado)
        df['SalePrice'] = np.log(df['SalePrice'])
        
        # plot
        plt.scatter(df[var], df['SalePrice'])
        plt.ylabel('SalePrice')
        plt.xlabel(var)
        plt.show()


for var in cont_vars:
    if var != 'SalePrice':
        transform_analyse_continuous(data, var)
```

De los gráficos anteriores, observamos algunas asociaciones monótonas entre SalePrice y las variables a las que aplicamos la transformación logarítmica, por ejemplo, 'GrLivArea'.

#### Outliers

Los valores extremos pueden afectar el rendimiento de un modelo lineal. Averigüemos si tenemos alguna en nuestras variables.

```python
# hagamos diagramas de caja para visualizar valores atípicos en las variables continuas


def find_outliers(df, var):
    df = df.copy()

    # log no toma valores negativos, así que seamos
    # cuidado y omite esas variables
    if any(data[var] <= 0):
        pass
    else:
        df[var] = np.log(df[var])
        df.boxplot(column=var)
        plt.title(var)
        plt.ylabel(var)
        plt.show()


for var in cont_vars:
    find_outliers(data, var)
```

La mayoría de las variables continuas parecen contener valores atípicos. Los valores atípicos tienden a afectar el rendimiento del modelo lineal. Por lo tanto, vale la pena dedicar un tiempo a comprender si la eliminación de valores atípicos agregará valor de rendimiento a nuestro modelo final de aprendizaje automático.

Vamos a evaluar cuántas categorías diferentes están presentes en cada una de las variables.

```python
data[cat_vars].nunique()
```

Todas las variables categóricas muestran baja cardinalidad, esto significa que tienen pocas etiquetas diferentes. Eso es bueno ya que no necesitaremos abordar la cardinalidad durante nuestra lección de ingeniería de características.

#### Etiquetas raras:

Avancemos e investiguemos ahora si hay etiquetas que están presentes solo en una pequeña cantidad de casas:

```python
def analyse_rare_labels(df, var, rare_perc):
    df = df.copy()

    # determine the % of observations por categoria
    tmp = df.groupby(var)[var].count() / len(df)

    # returna las categories que son rare
    return tmp[tmp < rare_perc]

# categorías de impresión que están presentes en menos de
# 1 % de las observaciones


for var in cat_vars:
    print(analyse_rare_labels(data, var, 0.01))
    print()
```

Algunas de las variables categóricas muestran etiquetas múltiples que están presentes en menos del 1% de las viviendas. Diseñaremos estas variables en nuestro próximo video. Las etiquetas que están subrepresentadas en el conjunto de datos tienden a provocar un ajuste excesivo de los modelos de aprendizaje automático. Por eso queremos eliminarlos.
