## Transformacion de variables Categoricas a numericas en Python

```python
#-variables ordinales: niveles de educación. Sus valores indican el nivel de presencia, o ausencia, de un aspecto concreto
#-Variable nominal: genero. nominales indican una categoría
```

#### 1. Carga de librerias y datos

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from matplotlib import rcParams
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
clientes=pd.read_csv('C:/Users/pc/Desktop/rociochavez/Rocio-Chavez-youtube-Files-master/Clientes Creado3.csv', index_col=0)
```

```python
clientes
```

```python
clientes.info()
```

### 1. Si todas las variables categoricas fueran nominales

```python
clientes_categ=clientes.select_dtypes(include=['object']).columns
clientes_categ
```

#### 2. Indicamos la transformación que queremos que se lleve a cabo

```python
#Creamos variables dummy

t=[('nominales', OneHotEncoder(drop='if_binary'), clientes_categ)]
```

#### 3. Aplicamos las transformaciones

```python
col_transform=ColumnTransformer(transformers=t)
col_transform
```

```python
clientes_numericas=col_transform.fit_transform(clientes)
clientes_numericas
```

```python
clientes_nombres=col_transform.get_feature_names_out()
clientes_nombres
```

```python
col_transform=ColumnTransformer(transformers=t, remainder='passthrough')
```

```python
clientes_numericas=col_transform.fit_transform(clientes)
clientes_numericas
```

```python
clientes_numericas_df=pd.DataFrame(clientes_numericas)
clientes_numericas_df
```

```python
clientes_numericas_df.columns=['Mujer', '1 a 3 Hijos', '3 o mas hijos', 'Ningun hijo', 'Licenciatura', 'Maestria', 'Preparatoria','Ingreso', 'edad']
clientes_numericas_df
```

### 2. Si tenemos variables nominales y Ordinales

```python
#Tomamos las variables genero y estudios como nominales y las variables hijos como ordinal

tr_nom=ColumnTransformer([('nominales', OneHotEncoder(drop='if_binary'), ['Genero', 'Estudios'])], remainder='passthrough')
```

```python
clientes_numericas2=tr_nom.fit_transform(clientes)
clientes_numericas2
```

```python
tr_nom.get_feature_names_out()
```

```python
clientes_numericas_df2=pd.DataFrame(clientes_numericas2)
clientes_numericas_df2
```

#### 2.1 Creamos los codificadores de las variables ordinales con un orden menor a mayor importancia

```python
from sklearn.preprocessing import OrdinalEncoder

tr_ord=ColumnTransformer([('hijos', OrdinalEncoder(categories=[('Mas de 3', '1 a 3', 'Ninguno')]), [4])],remainder='passthrough')
```

```python
clientes_numericas2=tr_ord.fit_transform(clientes_numericas_df2)
clientes_numericas2
```

```python
#Ejemplo de codigo si tuvieras mas de una variable ordinal


from sklearn.preprocessing import OrdinalEncoder

tr_ord2=ColumnTransformer([('hijos',  OrdinalEncoder(categories=[('Mas de 3', '1 a 3', 'Ninguno')]), ['Hijos']),
                          ('estudios', OrdinalEncoder(categories=[('Preparatoria', 'Licenciatura', 'Maestria')]), ['Estudios'])],
                          remainder='passthrough')
clientes_numericas4=tr_ord2.fit_transform(clientes)
clientes_numericas4
```
