```python
import numpy as np
data_points= np.random.normal(0,1, 100)
data_points
```

```python
plt.hist(data_points)
```

```python
import pylab
import scipy.stats as stats
stats.probplot (data_points, dist="norm", plot=pylab)
pylab.show()
```

```python
from scipy.stats import shapiro
estadistico, p_value = shapiro(data_points)
print("Estadistico= %.3f, p_vlue=%.3f" % (estadistico, p_value))
```

```python
plt.hist(contamina["Velocidad_viento"], edgecolor='black', linewidth=1)
```
